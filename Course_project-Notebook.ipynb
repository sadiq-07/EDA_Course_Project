```python
# Define the sleep disorder categories
sleep_disorders = ['Insomnia', 'Sleep Apnea', 'No Disorder']

# Plotting
plt.figure(figsize=(8, 18))

# Iterate over each occupation
for i, (occupation, group) in enumerate(df.groupby('Category')):
    # Count the occurrences of each sleep disorder within the occupation
    disorder_counts = group['Sleep Disorder'].value_counts()
    
    # Sort the sleep disorder counts in the order of sleep_disorders
    disorder_counts = disorder_counts.reindex(sleep_disorders, fill_value=0)
    
    # Plot horizontal bar graph for the current occupation
    plt.subplot(len(df['Occupation'].unique()), 1, i + 1)
    plt.barh(disorder_counts.index, disorder_counts.values, color=['orange', 'red', 'green'])
    plt.xlabel('Count')
    plt.ylabel('Sleep Disorder')
    plt.title(f'Distribution of Sleep Disorders among {occupation} ({len(group)} individuals)')
    plt.gca().invert_yaxis()  # Invert y-axis to have the highest count at the top

plt.tight_layout()
plt.show()
```
